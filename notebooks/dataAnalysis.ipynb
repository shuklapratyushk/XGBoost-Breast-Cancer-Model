```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid", context="notebook", palette="muted")
```

```python
df = pd.read_csv("../data/processed/breast_cancer_clean.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Diagnosis", data=df)
plt.title("Target Distribution (0 = Benign, 1 = Malignant)")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()

benign = (df["Diagnosis"] == 0).mean() * 100
malignant = (df["Diagnosis"] == 1).mean() * 100
print(f"Benign: {benign:.2f}% | Malignant: {malignant:.2f}%")
```

```python
subset = ["Diagnosis", "Radius_mean", "Texture_mean", "Perimeter_mean", "Area_mean"]
sns.pairplot(df[subset], hue="Diagnosis", diag_kind="kde", corner=True)
plt.suptitle("Feature Relationships (subset)", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
corr_target = corr["Diagnosis"].drop("Diagnosis").sort_values(ascending=False)
corr_target.head(10)
```

```python
corr_target.to_csv("../outputs/figures/feature_correlations.csv")
```
